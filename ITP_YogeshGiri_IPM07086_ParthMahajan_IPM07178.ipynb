```python
# Defining grades 
def grade_from_pct(p):
    if p >= 75: return 'A'
    if p >= 60: return 'B'
    if p >= 50: return 'C'
    if p >= 40: return 'D'
    return 'F'

n = int(input("How many students? ")) #taking inputs for the (n) number of students for which we have to calculate results 

passed = []
passed_with_grace = []
backlogs = []

for sid in range(1, n+1): # loop for splitting the values taken as strings 
    marks = list(map(int, input(f"Enter 5 marks for student {sid} (space separated): ").split()))
    if len(marks) != 5:
        print("Please enter exactly 5 numbers. Try this student again.")
        sid -= 1
        continue 

    total = sum(marks) #defining the processes of calculation and the criteria of backlog
    percent = total / 5.0
    # count subjects with marks < 40
    backlog_count = sum(1 for m in marks if m < 40)

    # applying grace rule
    if backlog_count == 0:
        g = grade_from_pct(percent)
        passed.append((sid, marks, total, percent, g))
    elif backlog_count == 1 and percent >= 45:
        # cleared by grace
        g = grade_from_pct(percent)
        passed_with_grace.append((sid, marks, total, percent, g))
    else:
        backlogs.append((sid, marks, total, percent, backlog_count)) # using append function to mention all the details

# Printing results of the examination
print("\n***** PASSED *****")
for s in passed:
    print(f"ID {s[0]}: Marks={s[1]} Total={s[2]} Pct={s[3]:.2f}% Grade={s[4]}")

print("\n___PASSED WITH GRACE___")
for s in passed_with_grace:
    print(f"ID {s[0]}: Marks={s[1]} Total={s[2]} Pct={s[3]:.2f}% Grade={s[4]} (1 backlog cleared by grace)")

print("\n---BACKLOGS---")
for s in backlogs:
    print(f"ID {s[0]}: Marks={s[1]} Total={s[2]} Pct={s[3]:.2f}% Backlogs={s[4]}")
```
